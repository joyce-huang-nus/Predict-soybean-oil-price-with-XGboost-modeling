```python
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import random
import datetime as dt
import statsmodels.api as sm

df_f = pd.read_csv(r'C:\Users\Joyce Huang\Downloads\Innovation challenge\final-financial.csv')
df_s = pd.read_csv(r'C:\Users\Joyce Huang\Downloads\Innovation challenge\soybean-oil-price.csv')
df_n = pd.read_csv(r'C:\Users\Joyce Huang\Downloads\Innovation challenge\news.csv')
df_s = df_s.dropna()
print(type(df_f['Date'][1]))
df_f['Date'] = pd.to_datetime(df_f['Date'])
df_s['Date'] = pd.to_datetime(df_s['Date'])
df_n['Date'] = pd.to_datetime(df_n['publish_time'])
print(type(df_f['Date'][1]))
print(type(df_s['Date'][1]))
print(type(df_n['Date'][1]))
df_s = df_s.sort_values(by=['Date'], ascending=True).reset_index()
display(df_f)
display(df_s)
display(df_n)
```

```python
merge = pd.merge(df_f, df_s, on ='Date', how ="outer")
merge = merge.drop('index',axis=1)
merge = pd.merge(merge, df_n, on ='Date', how ="outer")
merge = merge.drop(['Unnamed: 0','news_content','publish_time','year','Open','news_title'],axis=1)
merge = merge.sort_values(by=['Date'], ascending=True)
merge = merge[merge['Date'].dt.strftime('%Y')=='2021']

merge['Vol'] = merge['Vol'].str.replace('K', '')
merge['change'] = merge['Change %'].str.replace('%', '')
merge = merge.drop('Change %',axis=1)
display(merge)

e = ['Date','polarity','subjectivity']
df2 = merge[e]
```

```python
df = merge[['sp500','nadq','crude_oil','dow_jones','usd','Price','High','Low','Vol','change']]
from sklearn.impute import KNNImputer
import pandas as pd

imputer = KNNImputer()
imputed_data = imputer.fit_transform(df)
df_temp = pd.DataFrame(imputed_data)
df_temp.columns = df.columns
df['sp500'] = df_temp['sp500']
df['nadq'] = df_temp['nadq']
df['crude_oil'] = df_temp['crude_oil']
df['dow_jones'] = df_temp['dow_jones']
df['Price'] = df_temp['Price']
df['High'] = df_temp['High']
df['Low'] = df_temp['Low']
df['Vol'] = df_temp['Vol']
df['change'] = df_temp['change']
df['usd'] = df_temp['usd']
display(df)
```

```python
df = pd.concat([df,df2], axis=1)
df = df[['Date','sp500','nadq','crude_oil','dow_jones','usd','Price','High','Low','Vol','change','polarity','subjectivity']]
df = df.fillna(0)
df
```

```python
df.to_csv(r'C:\Users\Joyce Huang\Downloads\Innovation challenge\financial-soy-news.csv')
```

```python
from sklearn.model_selection import train_test_split,cross_validate, GridSearchCV
from sklearn.metrics import roc_curve, auc,recall_score,precision_score,plot_roc_curve,accuracy_score
import xgboost as xgb

X = df.drop(['Date','Price','High','Low','Vol'],axis=1)
Y = df['Price']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

def algorithm_pipeline(X_train, X_test, y_train, y_test, 
                       model, param_grid, cv=5, scoring_fit='neg_mean_squared_error',
                       do_probabilities = False):
    gs = GridSearchCV(
        estimator=model,
        param_grid=param_grid, 
        cv=cv, 
        n_jobs=-1, 
        scoring=scoring_fit,
        verbose=2
    )
    fitted_model = gs.fit(X_train, y_train)
    
    if do_probabilities:
      pred = fitted_model.predict_proba(X_test)
    else:
      pred = fitted_model.predict(X_test)
    
    return fitted_model, pred

model = xgb.XGBRegressor()
param_grid = {
    'learning_rate': (0.05, 0.10, 0.15),
    'n_estimators': [100, 150,200,250,300],
    'max_depth': [15,20,30,40,50]
}

model, pred = algorithm_pipeline(X_train, X_test, y_train, y_test, model, 
                                 param_grid, cv=5)

# Root Mean Squared Error
print(np.sqrt(-model.best_score_))
print(model.best_params_)
```
